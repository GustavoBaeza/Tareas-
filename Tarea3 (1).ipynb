```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.read_csv("https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/hdi.csv")
HDI=pd.read_csv("https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/hdi.csv")
pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto80/vacunacion_comuna_1eraDosis.csv")
dosis=pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto80/vacunacion_comuna_1eraDosis.csv")
```

```python
HDI
```

```python
RM=dosis[dosis['Region']=='Metropolitana']
RM
```

```python
col_list=list(RM)
col_list.remove('Codigo region')
col_list.remove('Codigo comuna')
col_list.remove('Poblacion')
col_list.remove('Region')
col_list.remove('Comuna')



RM=dosis[dosis["Region"].str.contains("Metropolitana")]
RM=RM.drop([140],axis=0)



RM["Gente_Vacunada"] = RM[col_list].sum(axis=1)
RM
```

```python
RM=RM[['Region','Comuna','Codigo comuna','Poblacion','Gente_Vacunada']]
RM
```

```python
HDI.rename(columns = {'comuna' : 'Codigo comuna'}, inplace = True)
```

```python
A=pd.merge(RM,HDI,
how='inner', on='Codigo comuna')
```

```python
from pylab import *
x1 = A['HDI']
y1 = A['Gente_Vacunada']
x= np.array(x1)
y= np.array(y1)
m,b=polyfit(x,y,1)
plot(x,y,'yo', x, m*x+b, '--k')
show()
```
