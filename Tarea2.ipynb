```python
col_list=list(RM)
col_list.remove('Codigo region')
col_list.remove('Codigo comuna')
col_list.remove('Poblacion')
col_list.remove('Region')
col_list.remove('Comuna')



RM=pdosis[pdosis["Region"].str.contains("Metropolitana")]
RM=RM.drop([140],axis=0)



RM["Total"] = RM[col_list].sum(axis=1)
RM
```

```python
#Se eliminan las columnas que no aportan a lo que se busca
RM=RM[['Region','Comuna','Codigo comuna','Poblacion','Total']]
RM
```

```python
L2['EDAD_0A5'] = L2['EDAD_0A5'].replace('*', 0)
L2['EDAD_6A14'] = L2['EDAD_6A14'].replace('*', 0)
L2['EDAD_15A64'] = L2['EDAD_15A64'].replace('*', 0)
L2['EDAD_65YMAS'] = L2['EDAD_65YMAS'].replace('*', 0)
L2
```

```python
#Se busca traspasar todo a numéricos para así poder hacer la suma en el paso siguiente 
type(L2['EDAD_0A5'][4])
L2['EDAD_0A5'] = pd.to_numeric(L2['EDAD_0A5'])
L2['EDAD_6A14'] = pd.to_numeric(L2['EDAD_6A14'])
L2['EDAD_15A64']= pd.to_numeric(L2['EDAD_15A64'])
L2['EDAD_65YMAS']= pd.to_numeric(L2['EDAD_65YMAS'])
L1.rename(columns = {'comuna' : 'Codigo comuna'}, inplace = True)
L2.rename(columns = {'COMUNA' : 'Codigo comuna'}, inplace = True)
L2
```

```python
L2=L2.groupby('Codigo comuna').sum()
```
