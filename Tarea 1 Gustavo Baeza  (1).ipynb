```python
import random
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy
import csv

archivo = open('FallecidosEtario_T.csv') 
reader = pd.read_csv(archivo)
reader = reader.values
F=[]
F2=[]
M=[]
n=0
l = []

for row in reader:
    n=n+1
    x=row[0]
    y=int(row[1])+int(row[2])+int(row[3])+int(row[4])+int(row[5])+int(row[6])+int(row[7]) 
    F.append(x)
    l.append(y)
    
    if n/7==1: 
        F2.append(x)
        M.append(y)
        n=0

fig, ax = plt.subplots()
ax.plot(F, l, linestyle = "dashed")
plt.show()         
        
M3 = []

BD=0

for i in range(len(l)):
    FD=l[i]-l[i-1]
    M3.append(BD)
    
M3[0]=l[0]


fig, ax = plt.subplots()
ax.plot(F, M3, linestyle = "dashed")
plt.show()  

        

fig, ax = plt.subplots()
ax.plot(F2, M, linestyle = "dashed")
plt.show()  






M2 = []
FD = 0
for i in range(len(M)):
    FD=M[i]-M[i-1]
    M2.append(FD)
    
M2[0]=M[0]
    
fig, ax = plt.subplots()
ax.plot(F2, M2, linestyle = "dashed")
plt.show()  


```
